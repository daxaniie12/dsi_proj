```python
from sklearn.feature_selection import SelectKBest, f_regression, f_classif
# Filtering numerical predictors

skb_n = SelectKBest(f_regression)
skb_n.fit(numerical_f_df, house_f['SalePrice'])
# examine results
kbest_n = pd.DataFrame([numerical_f_df.columns, list(skb_n.scores_), list(skb_n.pvalues_)], index=['feature','f_reg_score', 'f_reg_pvalue']).T.sort_values('f_reg_pvalue', ascending=False)
kbest_n
```

```python
#Getting a list of numerical features significantly correlated with target variable 'SalePrice'
num_f = []
for i,j in zip(kbest_n['feature'], kbest_n['f_reg_pvalue']):
    if j <= 0.05:
        num_f.append(i)
print "Numerical features significantly correlated with target variable 'SalePrice':"
print num_f
print len(num_f)
```

```python
# Import relevant modules for modeling
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV
from sklearn import metrics
```

```python
#Getting the training and test dataframe, train is pre-2010, test is 2010 for 'YrSold'
house_fixed_train = house_fixed[house_fixed['YrSold'] != 2010]
house_fixed_test = house_fixed[house_fixed['YrSold'] == 2010]
```

```python
non_predictors = ['SalePrice_Ks', 'YrSold']
predictor_cols = [i for i in house_fixed.columns if i not in non_predictors]
```

```python
# Defining X_train, X_test, y_train and y_test
X_train = house_fixed_train.loc[:, predictor_cols]
X_test = house_fixed_test.loc[:, predictor_cols]
y_train = house_fixed_train['SalePrice_Ks']
y_test = house_fixed_test['SalePrice_Ks']
```

```python
#Let's try fitting 3 different linear models
#One Lasso, one Ridge, one SGDRegressor
```

```python
#RidgeCV
ridge_alphas = np.logspace(0, 5, 200)

ridge = RidgeCV(alphas=ridge_alphas, cv=10)
ridge.fit(X_train, y_train)
y_pred1 = ridge.predict(X_test)
print 'R2 value:', ridge.score(X_test, y_test)
```

```python
#LassoCV
lasso = LassoCV(n_alphas=500, cv=10, verbose=1)
lasso.fit(X_train, y_train)
y_pred2 = lasso.predict(X_test)
print 'R2 value:', lasso.score(X_test, y_test)
```

```python
#ElasticNetCV
l1_ratios = np.linspace(0.01, 1.0, 25)
enet = ElasticNetCV(l1_ratio=l1_ratios, n_alphas=100, cv=10,
                            verbose=1)
enet.fit(X_train, y_train)
ypred3 = enet.predict(X_test)
print 'R2 value:', enet.score(X_test, y_test)
```

```python
#RidgeCV gives the best R2 value, but I will use the LassoCV model since I wanr to do feature selection
```

```python
lasso_coefs = pd.DataFrame({'variable':X_train.columns,
                            'coef':lasso.coef_,
                            'abs_coef':np.abs(lasso.coef_)})

lasso_coefs.sort_values('abs_coef', inplace=True, ascending=False)

lasso_coefs
```

```python
useful_cols = [j for i,j in zip(lasso_coefs['abs_coef'],lasso_coefs['variable']) if i!=0]
print useful_cols
```
