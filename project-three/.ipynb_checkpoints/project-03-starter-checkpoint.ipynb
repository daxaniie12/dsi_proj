```python
# Import relevant modules
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV, Ridge, Lasso, ElasticNet
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
#Getting the training and test dataframe, train is pre-2010, test is 2010 for 'YrSold'
house_fixed_train = house_fixed[house_fixed['YrSold'] != 2010]
house_fixed_test = house_fixed[house_fixed['YrSold'] == 2010]
```

```python
# Defining X_train, X_test, y_train and y_test
X_train = house_fixed_train.loc[:, predictor_cols]
X_test = house_fixed_test.loc[:, predictor_cols]
y_train = house_fixed_train['SalePrice_Ks']
y_test = house_fixed_test['SalePrice_Ks']
```

```python
#Let's try fitting 3 different linear models
#One Lasso, one Ridge, one ElasticNet
```

```python
ls = LinearRegression()
ls.fit(X_train, y_train)
R2_score = ls.score(X_train, y_train)
y_pred = ls.predict(X_test)
mse1 = metrics.mean_squared_error(y_test, y_pred)
print 'R-squared:', R2_score
print 'Mean squared error:', mse1
```

```python
#RidgeCV
ridge_alphas = np.logspace(0, 5, 200)

optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(X_train, y_train)

print optimal_ridge.alpha_
```

```python
ridge = Ridge(alpha=optimal_ridge.alpha_)

ridge_scores = cross_val_score(ridge, X_train, y_train, cv=10)

print ridge_scores
print np.mean(ridge_scores)
```

```python
#LassoCV
optimal_lasso = LassoCV(n_alphas=500, cv=10, verbose=1)
optimal_lasso.fit(X_train, y_train)

print optimal_lasso.alpha_
```

```python
lasso = Lasso(alpha=optimal_lasso.alpha_)

lasso_scores = cross_val_score(lasso, X_train, y_train, cv=10)

print lasso_scores
print np.mean(lasso_scores)
```

```python
#ElasticNetCV
l1_ratios = np.linspace(0.01, 1.0, 25)

optimal_enet = ElasticNetCV(l1_ratio=l1_ratios, n_alphas=100, cv=10,
                            verbose=1)
optimal_enet.fit(X_train, y_train)

print optimal_enet.alpha_
print optimal_enet.l1_ratio_
```

```python
enet = ElasticNet(alpha=optimal_enet.alpha_, l1_ratio=optimal_enet.l1_ratio_)

enet_scores = cross_val_score(enet, X_train, y_train, cv=10)

print enet_scores
print np.mean(enet_scores)
```

```python
#Fitting model with Lasso since I want to do feature selection
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=211.938958483)
lasso.fit(X_train, y_train)
```

```python
useful_cols = [j for i,j in zip(lasso_coefs['abs_coef'],lasso_coefs['variable']) if i!=0]
print useful_cols
```

Now that you have a model that estimates the price of a house based on its static characteristics, we can move forward with part 2 and 3 of the plan: what are the costs/benefits of quality, condition, and renovations?
