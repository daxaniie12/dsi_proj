```python
#Convert codes in diag columns to respective diagnosis labels
```

```python
#Create lists containing respective codes per diagnosis type
#start and stop refer to the starting and ending numbers of the range of codes
#ind refers to individual code per disease type
def floatnum(start,stop,ind=None):
    num = range(start,stop+1,1)
    if ind != None:
        num.append(ind)
    num = map(float, num)
    return num
```

```python
Circulatory = floatnum(390,459,785)
Respiratory = floatnum(460, 519, 786)
Digestive = floatnum(520,579,787)
Injury = floatnum(800,999)
Musculoskeletal = floatnum(710,739)
Genitourinary = floatnum(580,629,788)
Neoplasms = floatnum(140,239)
```

```python
# Convert the codes in 'diag_1', 'diag_2', 'diag_3' based on the above-mentioned codes
def convert(data):
    if data in Circulatory:
        return 'Circulatory'
    elif data in Respiratory:
        return 'Respiratory'
    elif data in Digestive:
        return 'Digestive'
    elif str(data)[:3] == '250':
        return 'Diabetes'
    elif data in Injury:
        return 'Injury'
    elif data in Musculoskeletal:
        return 'Musculoskeletal'
    elif data in Genitourinary:
        return 'Genitourinary'
    elif data in Neoplasms:
        return 'Neoplasms'
    else:
        return 'Others'
```

```python
diag_1 = diabetes['diag_1'].apply(convert)
diag_2 = diabetes['diag_2'].apply(convert)
diag_3 = diabetes['diag_3'].apply(convert)
```

```python
class Swarmplots(object):
    def __init__(self, feature, target, df):
        self.feature = feature
        self.target = target
        self.df = df
        
    def outliers(self,upper=75,lower=25,IQRM=1.5):
        IQR = np.percentile(np.array(self.df[self.feature]), upper) - np.percentile(np.array(self.df[self.feature]), lower)
        self.IQR_15 = np.percentile(np.array(self.df[self.feature]), upper) + IQR*IQRM
        dfo = self.df[[self.feature,self.target]]
        self.dfo = dfo[dfo[self.feature]>self.IQR_15]
        return self.dfo, self.IQR_15
    
    def normal(self):
        dfn = self.df[[self.target, self.feature]]
        self.dfn = dfn[dfn[self.feature]<=self.IQR_15]
        return self.dfn
        
    def sample(self, size=1000):
        np.random.seed(12)
        ran_indices = np.random.choice(self.dfn.index, size, replace=False)
        self.df_ran = self.dfn.ix[ran_indices,:]
        if self.dfo.shape[0] > size:
            ran_indiceso = np.random.choice(self.dfo.index, size, replace=False)
            self.df_ran_o = self.dfo.ix[ran_indiceso,:]
        else:
            self.df_ran_o = self.dfo.copy()
        return self.df_ran, self.df_ran_o
        
    def splot(self):
        fig, axes = plt.subplots(1,2, figsize=(14,6))
        norm = sns.swarmplot(y=self.feature, x=self.target, palette='husl', data= self.df_ran, ax=axes[0])
        out =  sns.swarmplot(y=self.feature, x=self.target, palette='husl', data= self.df_ran_o, ax=axes[1])
        return norm, out
```

```python
#Plot on the left is target class distribution of 1000 randomly selected datapoints from df that are not outliers
#Plot on the right is target class distribution of outlier points
```

```python
#Get the target class count for each numerical outliers feature
def outliers_dist(dfs):
    dist = pd.DataFrame()
    for i in dfs:
        #print i
        count = i.groupby('readmitted').count().reset_index()
        name = list(count.columns)[1]
        #print name
        x = count.iloc[:,1]
        dist[name] = count.iloc[:,1]/np.sum((x[0],x[1],x[2]))
    return dist
```

```python
class Catplots(object):
    
    def __init__(self, feature, target, df):
        self.feature = feature
        self.target = target
        self.df = df
    
    '''def sample(self, size=1000):
        np.random.seed(12)
        ran_indices = np.random.choice(self.df.index, size, replace=False)
        self.df_ran = self.df.ix[ran_indices,:]
        return self.df_ran'''
    
    def countplot(self):
        return sns.countplot(self.feature, hue=self.target, data=self.df, palette='husl')
    
    def barplot(self):
        x = pd.crosstab(self.df[self.target], self.df[self.feature])
        y = relative = x.div(x.sum(axis=0), axis=1)
        y = relative.T.plot(kind='bar')
        plt.ylabel('proportion')
        return y
```

```python
#Run chi-square test of independence between 'max_glu_serum' and 'A1Cresult'
# Chi-sqaure contingency table between 'max_glu_serum' and 'A1Cresult'
#Contingency table is the same as cross-tabulation

import scipy.stats as stats

cont = pd.crosstab(diabetes4['A1Cresult'], diabetes4['max_glu_serum'])
results = stats.chi2_contingency(cont)
chi2 = results[0]
p_val = results[1]
print "chi-square value:", chi2
print "p-value:", p_val
```

Since glucose serum test and A1C tests are both tests for the level of plasma glucose, the chisquare test of independence was carried out to see if these variables are independent. Since p-value<0.01, these 2 variables are not independent, i.e. they are associated
