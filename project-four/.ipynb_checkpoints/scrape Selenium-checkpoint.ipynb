```python
from bs4 import BeautifulSoup
import time
from selenium import webdriver
import pandas as pd
import numpy as np
```

```python
chrome_path = './chromedriver/chromedriver'
driver = webdriver.Chrome(chrome_path)
driver.get('https://myjobstreet.jobstreet.com.sg/home/login.php?site=sg&language_code=3')
```

```python
def scrape_jobstreet(job_names, pages=None):
    jobs_df = []
    companies_df = []
    locations_df = []
    salaries_df = []
    industries_df = []
    experiences_df = []
    company_sizes_df = []
    descriptions_df = []
    
    def jjobs(info):
        jobs = []
        companies = []
        locations = []
        salaries = []
        linkss = []
        for inf in info:
            job = inf.findAll('h2', {'itemprop':'title'})
            for j in job:
                jobs.append(j.text)
            company = inf.findAll('span', {'itemprop':'name'})
            for com in company:
                companies.append(com.text)
            location = inf.findAll('span', {'itemprop':'jobLocation'})
            for loc in location:
                locations.append(loc.text)
            salary = inf.findAll('font')
            for sal in salary:
                salaries.append(sal.text.strip())
        # print sal.get('class')
            links = inf.findAll('a', {'class', 'position-title-link'})
            for link in links:
                linkss.append(link.get('href'))
        return jobs, companies, locations, salaries, linkss
    
    def get_info(linkss):
        industries = []
        experiences = []
        company_sizes =[]
        descriptions = []
        for link in linkss:
            driver.get(link)
            source = driver.page_source
            soup = BeautifulSoup(source, "html.parser")
            industries.append(soup.findAll('p', {'id':'company_industry'})[0].text)
            experiences.append(soup.findAll('span', {'itemprop':'experienceRequirements'})[0].text.strip())
            company_sizes.append(soup.findAll('p', {'id':'company_size'})[0].text.strip())
            descriptions.append(soup.findAll('div', attrs = {'id':'job_description'})[0].text)
        return industries, experiences, company_sizes, descriptions 
    
    for job in job_names:
        for pag in range(0, pages+1, 1):
            job = str(job).split()
            url = 'https://www.jobstreet.com.sg/en/job-search/job-vacancy.php?key=' + str(job[0]) + '+' + str(job[1]) + '&area=1&option=1&job-source=1%2C64&classified=1&job-posted=0&sort=2&order=0&pg=' + str(pag) + '&src=16&ojs=2' 
            driver.get(url)
            source = driver.page_source
            soup = BeautifulSoup(source, "html.parser")
            info = soup.findAll('div', {'class':'panel-body'})
            time.sleep(3)
    
            jobs_df.extend(jjobs(info)[0])
            companies_df.extend(jjobs(info)[1])
            locations_df.extend(jjobs(info)[2])
            salaries_df.extend(jjobs(info)[3])
            #industries_df.extend(get_info((jjobs[4])[0]))
            #experiences_df.extend(get_info((jjobs[4])[1]))
            #company_sizes_df.extend(get_info((jjobs[4])[2]))
            #descriptions_df.extend(get_info((jjobs[4])[3]))
    
    
    jobstreet = pd.DataFrame()
    jobstreet['job_title'] = jobs_df
    jobstreet['company'] = companies_df
    jobstreet['location'] = locations_df
    jobstreet['salary'] = salaries_df
    #jobstreet['summary'] = summaries_df
    
    return jobstreet
```

```python
import numpy as np
```

```python
def jjobs(info):
    jobs = []
    companies = []
    locations = []
    linkss = []
    for inf in info:
        job = inf.findAll('h2', {'itemprop':'title'})
        for j in job:
            jobs.append(j.text)
        company = inf.findAll('span', {'itemprop':'name'})
        for com in company:
            companies.append(com.text)
        location = inf.findAll('span', {'itemprop':'jobLocation'})
        for loc in location:
            locations.append(loc.text)
        links = inf.findAll('a', {'class', 'position-title-link'})
        for link in links:
            linkss.append(link.get('href'))
    return jobs, companies, locations, linkss
```

```python
def get_info(linkss):
        salaries = []
        industries = []
        experiences = []
        company_sizes =[]
        descriptions = []
        for link in linkss:
            driver.get(link)
            source = driver.page_source
            soup = BeautifulSoup(source, "html.parser")
            try:
                sal = soup.findAll('span', {'id':'salary_range'})[0].text.strip()
                salaries.append(sal)
            except:
                salaries.append(np.nan)
            try:
                ind = soup.findAll('p', {'id':'company_industry'})[0].text
                industries.append(ind)
            except:
                industries.append(np.nan)
            try:
                exp = soup.findAll('span', {'itemprop':'experienceRequirements'})[0].text.strip()
                experiences.append(exp)
            except:
                experiences.append(np.nan)
            try:
                com = soup.findAll('p', {'id':'company_size'})[0].text.strip()
                company_sizes.append(com)
            except:
                company_sizes.append(com)
            try:
                des = soup.findAll('div', attrs = {'id':'job_description'})[0].text
                descriptions.append(des)
            except:
                descriptions.append(np.nan)
            time.sleep(3)
        return salaries, industries, experiences, company_sizes, descriptions 
```

```python
def dataframe(jobs, companies, locations, salaries, industries, experiences, company_sizes, descriptions):
    df = pd.DataFrame()
    df['job_title'] = jobs
    df['company'] = companies
    df['location'] = locations
    df['salary'] = salaries
    df['industry'] = industries
    df['experience'] = experiences
    df['company_size'] = company_sizes
    df['job_description'] = descriptions
    return df
```
