```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

```python
locationss = ['New York', 'Chicago', 'San Francisco', 'Austin', 'Seattle',
                  'Los Angeles', 'Philadelphia', 'Atlanta', 'Dallas',
                  'Pittsburgh', 'Portland', 'Phoenix', 'Denver', 'Houston','Miami']
```

```python
def scrape_indeed(job_names, locationss, pages=None):
    
    jobs_df = []
    companies_df = []
    locations_df = []
    salaries_df = []
    summaries_df = []
    
    def ijob_title(data): 
        jobs = []
        for div in soup.find_all(name='div', attrs={'class':'row'}):
            for a in div.find_all(name='a', attrs={'data-tn-element':'jobTitle'}):
                jobs.append(a['title'])
        return jobs
    def icompany(soup): 
        companies = []
        for div in soup.find_all('div', attrs={'class':'row'}):
            company = div.find_all('span', attrs={'class':'company'})
            if len(company) > 0:
                for b in company:
                    companies.append(b.text.strip())
            else:
                sec_try = div.find_all('span', attrs={'class':'result-link-source'})
                for span in sec_try:
                    companies.append(span.text.strip())
        return companies
 
    def ilocation(data): 
        locations = []
        spans = soup.findAll('span', attrs={'class': 'location'})
        for span in spans:
            locations.append(span.text)
        return locations
    
    def isalary(data): 
        salaries = []
        listings = soup.findAll('div',{'data-tn-component':'organicJob'})
        for listing in listings:
            try:
                salary = listing.find('td', class_='snip').find('span', class_='no-wrap').get_text().strip()
                salaries.append(salary)
            except:
                salaries.append('none')
                #salary = 'none'
            #salaries.append(salary)
        return salaries
        '''for div in soup.find_all(name='div', attrs={'class':'row'}):
            try:
                salaries.append(div.find('nobr').text)
            except:
                try:
                    div_two = div.find(name='div', attrs={'class':'sjcl'})
                    div_three = div_two.find('div')
                    salaries.append(div_three.text.strip())
                except:
                    salaries.append('None')'''
       
    def isummary(data): 
        summaries = []
        spans = soup.findAll('span', attrs={'class': 'summary'})
        for span in spans:
            summaries.append(span.text.strip())
        return summaries
    
    for job in job_names:
        for loc in locationss:
            for pag in range(0, pages*10+1, 10):
                job = str(job).split()
                #url = https://www.indeed.com/jobs?q=data+analyst&l=San+Francisco%2C+CA&start=20
                url = 'https://www.indeed.com/jobs?q=' + str(job[0]) + '+' + str(job[1]) + '&l=' + loc.replace(' ','+') + '&start=' + str(pag)
                response = requests.get(url)
                html = response.text
                soup = BeautifulSoup(html, 'html.parser')
                time.sleep(1)
    
                jobs_df.extend(ijob_title(soup))
                companies_df.extend(icompany(soup))
                locations_df.extend(ilocation(soup))
                salaries_df.extend(isalary(soup))
                summaries_df.extend(isummary(soup))
    
    indeed = pd.DataFrame()
    indeed['job_title'] = jobs_df
    indeed['company'] = companies_df
    indeed['location'] = locations_df
    indeed['salary'] = salaries_df
    indeed['summary'] = summaries_df
    
    return indeed
```
